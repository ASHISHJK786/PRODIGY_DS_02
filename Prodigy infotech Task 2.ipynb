# What is EDA

What is EDA (Exploratory Data Analysis) in Data Analysis?
Exploratory Data Analysis (EDA) is a crucial step in the data analysis process, where the primary goal is to examine the data and summarize its main characteristics, often using visual methods. It is a method to understand the underlying patterns, relationships, and structures in the data before formal modeling or hypothesis testing. EDA helps data analysts, data scientists, and researchers to:

Understand the data (content, structure, and potential relationships).
Detect errors or outliers (values that don't fit expected patterns).
Make decisions about further analysis (e.g., selecting features for models).
Formulate hypotheses based on data insights.
Key Aspects of EDA
Data Summarization: EDA typically involves computing summary statistics to get a high-level understanding of the data, such as:

Mean, Median, Mode
Standard Deviation, Variance
Min, Max
Quartiles, Percentiles
Data Visualization: Visual representations help to see patterns or trends in the data. Some commonly used plots and charts include:

Histograms: For understanding distributions.
Box Plots: To find outliers and understand the spread.
Scatter Plots: To analyze relationships between two variables.
Bar Charts: To compare categorical data.
Heatmaps: To understand correlations between variables.
Detecting Missing Values and Outliers: During EDA, analysts often check for:

Missing data: Identify and handle missing or incomplete records.
Outliers: Detect and analyze unusual data points that could affect the overall analysis.
Data Distribution: EDA involves understanding how your data is distributed, which helps you decide the type of statistical or machine learning techniques that can be applied.

Normality: Checking whether the data follows a normal distribution.
Skewness: Whether the data is asymmetric.
Kurtosis: How "heavy-tailed" the data distribution is.
Data Relationships: EDA helps in understanding how different features or variables are related to one another.

Correlation Analysis: To measure the degree to which two variables move in relation to each other (correlation coefficient).
Covariance: Understanding how two variables change together.
Feature Engineering: EDA allows for new insights, which often result in creating new features or transformations of existing data for better model performance.

Importance of EDA
Data Quality Assessment: EDA helps to clean and prepare data by identifying missing values, inconsistencies, and outliers that might negatively affect the model.

Model Selection and Tuning: EDA can guide which machine learning algorithms or statistical methods are appropriate based on the data distribution and relationships between variables.

Preliminary Insights: EDA can give a quick overview of data characteristics, which can suggest potential hypotheses, trends, or patterns before proceeding with more formal and advanced analysis.

Better Decision Making: By exploring the data visually, you can make more informed decisions about feature selection, transformations, and preprocessing steps.

Tools for EDA
Pandas: For data manipulation and summary statistics in Python.
Matplotlib/Seaborn: For visualization in Python.
R: With libraries like ggplot2, dplyr, and tidyverse for both data wrangling and visualization.
Tableau/Power BI: For interactive, visual-based EDA.

# Visualisation

# 1. Survival Rate by Gender
Relationship: Gender and survival rates.
Visualization: Create a bar chart or a grouped bar chart showing 
the proportion of males vs. females who survived. 
This can provide insight into whether gender played a significant role in survival.


```python
# Load the train dataset
df_train = pd.read_csv("train.csv")

# Set the style of seaborn
sb.set(style="whitegrid")

# Create a bar chart showing survival rate by gender
mpl.figure(figsize=(8, 6))
sb.barplot(x='Sex', y='Survived', data=df_train)

# Add labels and title
mpl.xlabel('Gender')
mpl.ylabel('Survival Rate')
mpl.title('Survival Rate by Gender')


mpl.savefig('Survival Rate by Gender.png', dpi=300)


# Display the plot
mpl.show()
```

#  2 Survival Rate by Passenger Class (Pclass)
Relationship: Passenger class and survival.
Visualization: A bar chart or stacked bar
chart showing survival rates for each passenger class (Pclass). This helps identify whether passengers in higher classes had a better chance of survival.

```python
# Load the data from the train CSV file
df_train = pd.read_csv("train.csv")

# Calculate the survival rate for each Pclass
survival_rate_by_class = df_train.groupby('Pclass')['Survived'].mean() * 100

# Plot the survival rate by Pclass
mpl.figure(figsize=(8, 6))
mpl.bar(survival_rate_by_class.index, survival_rate_by_class, color=['skyblue', 'lightgreen', 'salmon'])

# Add labels and title
mpl.xlabel('Passenger Class (Pclass)', fontsize=12)
mpl.ylabel('Survival Rate (%)', fontsize=12)
mpl.title('Survival Rate by Passenger Class (Pclass)', fontsize=14)

# Show the percentage on the bars
for i, value in enumerate(survival_rate_by_class):
    mpl.text(i + 1, value + 1, f'{value:.2f}%', ha='center', fontsize=10)

    
mpl.savefig('Survival Rate by Passenger Class (Pclass).png', dpi=300)

# Display the plot
mpl.tight_layout()
mpl.show()
```

# 3. Age Distribution of Survivors and Non-Survivors
Relationship: Age and survival.
Visualization: A kernel density plot (KDE) or a histogram showing the distribution of ages for both survivors and non-survivors. You can also create a box plot to show how age might have impacted survival.

```python
# Load the data from the train CSV file
df_train = pd.read_csv("train.csv")

# Remove rows with missing values in the 'Age' and 'Survived' columns
df_train = df_train.dropna(subset=['Age', 'Survived'])

# Set plot style
sb.set(style="whitegrid")

# Create a figure with subplots
fig, ax = mpl.subplots(1, 2, figsize=(14, 6))

# --- Kernel Density Plot (KDE) ---
sb.kdeplot(df_train[df_train['Survived'] == 1]['Age'], ax=ax[0], shade=True, color='green', label='Survived')
sb.kdeplot(df_train[df_train['Survived'] == 0]['Age'], ax=ax[0], shade=True, color='red', label='Not Survived')

# Add labels and title for KDE plot
ax[0].set_title('Age Distribution: Survivors vs. Non-Survivors', fontsize=14)
ax[0].set_xlabel('Age', fontsize=12)
ax[0].set_ylabel('Density', fontsize=12)
ax[0].legend()

# --- Box Plot ---
sb.boxplot(x='Survived', y='Age', data=df_train, ax=ax[1], palette={'1': 'green', '0': 'red'})

# Add labels and title for Box Plot
ax[1].set_title('Age Distribution: Survivors vs. Non-Survivors (Box Plot)', fontsize=14)
ax[1].set_xlabel('Survived', fontsize=12)
ax[1].set_ylabel('Age', fontsize=12)
ax[1].set_xticklabels(['Not Survived', 'Survived'])

# Adjust layout for better fit
mpl.tight_layout()

# Save the figure
mpl.savefig('age_distribution_survivors_vs_non_survivors.png', dpi=300)

# Display the plots
mpl.show()
```

# 4. Survival by Number of Siblings/Spouses (SibSp) and Parents/Children (Parch)
Relationship: Family size and survival.
Visualization: Create a bar chart or heatmap to see how survival is influenced by the number of siblings/spouses (SibSp) or parents/children (Parch). Larger families may have had a lower chance of survival.

```python
# Load the dataset
df_train = pd.read_csv("train.csv")

# Remove rows with missing values in 'SibSp', 'Parch', and 'Survived' columns
df_train = df_train.dropna(subset=['SibSp', 'Parch', 'Survived'])

# Set style for seaborn plots
sb.set(style="whitegrid")

# Create a new DataFrame to calculate survival rates based on SibSp and Parch
survival_family = df_train.groupby(['SibSp', 'Parch'])['Survived'].mean().reset_index()

# Create a pivot table for the heatmap
heatmap_data = survival_family.pivot_table(index='SibSp', columns='Parch', values='Survived')

# Create a figure
mpl.figure(figsize=(12, 6))

# Create a heatmap for survival rates based on SibSp and Parch
sb.heatmap(heatmap_data, annot=True, cmap='coolwarm', cbar_kws={'label': 'Survival Rate'}, linewidths=0.5)

# Set labels and title
mpl.title('Survival Rate by Number of Siblings/Spouses and Parents/Children', fontsize=14)
mpl.xlabel('Number of Parents/Children (Parch)', fontsize=12)
mpl.ylabel('Number of Siblings/Spouses (SibSp)', fontsize=12)

# Save the figure
mpl.savefig('survival_by_family_size.png', dpi=300)

# Show the plot
mpl.show()
```

#  5. Embarked Location and Survival
Relationship: Port of embarkation and survival.
Visualization: A bar chart showing survival rates based on the port where passengers boarded (Embarked). This might reveal if certain ports had a higher survival rate.

```python
# Load the dataset
df_train = pd.read_csv("train.csv")

# Remove rows with missing values in 'Embarked' and 'Survived' columns
df_train = df_train.dropna(subset=['Embarked', 'Survived'])

# Set style for seaborn plots
sb.set(style="whitegrid")

# Calculate survival rates based on the port of embarkation
embarked_survival = df_train.groupby('Embarked')['Survived'].mean().reset_index()

# Create a bar plot for survival rates by port of embarkation
mpl.figure(figsize=(8, 6))
sb.barplot(x='Embarked', y='Survived', data=embarked_survival, palette='pastel')

# Set labels and title for the plot
mpl.title('Survival Rate by Port of Embarkation', fontsize=14)
mpl.xlabel('Port of Embarkation', fontsize=12)
mpl.ylabel('Survival Rate', fontsize=12)
mpl.ylim(0, 1)  # Set y-axis limits to represent percentage

# Save the figure
mpl.savefig('survival_by_embarked.png', dpi=300)

# Show the plot
mpl.show()
```

# 6. Survival Rate by Family Size
Relationship: Combine SibSp and Parch to create a new feature called "Family Size" and analyze its impact on survival.
Visualization: Bar plot or KDE to show the survival probability with respect to the size of the family.

```python
# Load the dataset
df_train = pd.read_csv("train.csv")

# Create a new feature for Family Size
df_train['Family Size'] = df_train['SibSp'] + df_train['Parch'] + 1  # +1 for the passenger themselves

# Set style for seaborn plots
sb.set(style="whitegrid")

# Calculate survival rates based on Family Size
family_survival = df_train.groupby('Family Size')['Survived'].mean().reset_index()

# Create a bar plot for survival rates by Family Size
mpl.figure(figsize=(12, 6))
sb.barplot(x='Family Size', y='Survived', data=family_survival, palette='pastel')

# Set labels and title for the plot
mpl.title('Survival Rate by Family Size', fontsize=14)
mpl.xlabel('Family Size', fontsize=12)
mpl.ylabel('Survival Rate', fontsize=12)
mpl.ylim(0, 1)  # Set y-axis limits to represent percentage

# Save the figure
mpl.savefig('survival_by_family_size.png', dpi=300)

# Show the plot
mpl.show()
```

#  7. Correlation Heatmap
Relationship: Correlation between all numerical variables.
Visualization: A heatmap to show correlations between the variables such as Age, Fare, Pclass, SibSp, Parch, etc.

```python
# Load the dataset
df_train = pd.read_csv("train.csv")

# Select numerical columns for correlation
numerical_columns = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

# Calculate the correlation matrix
correlation_matrix = df_train[numerical_columns].corr()

# Set style for seaborn plots
sb.set(style="whitegrid")

# Create a heatmap for the correlation matrix
mpl.figure(figsize=(10, 8))
sb.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar_kws={'label': 'Correlation Coefficient'}, linewidths=0.5)

# Set title for the heatmap
mpl.title('Correlation Heatmap of Numerical Variables', fontsize=14)

# Save the figure
mpl.savefig('correlation_heatmap.png', dpi=300)

# Show the plot
mpl.show()
```
